```python
import numpy as np
import matplotlib.pyplot as plt

# Continuity and Differentiation

# Define a function
def f(x):
    return np.sin(x)

# Define its derivative
def f_prime(x):
    return np.cos(x)

# Plot the function and its derivative
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, f_prime(x), label="f'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and its Derivative')
plt.legend()
plt.grid(True)
plt.show()

# Taylor Theorem

# Define a function to calculate the nth derivative of sin(x)
def nth_derivative_sin(n, x):
    remainder = n % 4
    if remainder == 0:
        return np.sin(x)
    elif remainder == 1:
        return np.cos(x)
    elif remainder == 2:
        return -np.sin(x)
    else:
        return -np.cos(x)

# Define the Taylor series approximation for sin(x)
def taylor_series_sin(n, x, a):
    result = 0
    for i in range(n+1):
        result += nth_derivative_sin(i, a) / np.math.factorial(i) * (x - a)**i
    return result

# Plot sin(x) and its Taylor series approximation at x=0 for different degrees
x_values = np.linspace(-2*np.pi, 2*np.pi, 100)
a = 0  # Expansion point
degrees = [1, 3, 5, 7]

plt.figure(figsize=(10, 6))
plt.plot(x_values, np.sin(x_values), label='sin(x)', color='blue')
for degree in degrees:
    plt.plot(x_values, taylor_series_sin(degree, x_values, a), label=f'Taylor Series (degree={degree})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Series Approximation of sin(x)')
plt.legend()
plt.grid(True)
plt.show()

# Optimization with Gradient Descent

# Define the function to be optimized
def f(x):
    return x**2 - 4*x + 4

# Define its derivative
def f_prime(x):
    return 2*x - 4

# Gradient descent optimization algorithm
def gradient_descent(f, f_prime, x_init, learning_rate, num_iterations):
    x_values = [x_init]
    for _ in range(num_iterations):
        x_new = x_values[-1] - learning_rate * f_prime(x_values[-1])
        x_values.append(x_new)
    return x_values

# Perform gradient descent optimization
x_init = 5  # Initial guess
learning_rate = 0.1
num_iterations = 20
x_values = gradient_descent(f, f_prime, x_init, learning_rate, num_iterations)

# Plot the function and the optimization path
x = np.linspace(0, 5, 100)
plt.plot(x, f(x), label='f(x)')
plt.scatter(x_values, f(np.array(x_values)), color='red', label='Optimization Path')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Optimization with Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
